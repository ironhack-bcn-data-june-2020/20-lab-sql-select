# Challenge 1

```python
%%sql
SELECT a.au_id,a.au_lname as last_name,a.au_fname as first_name,t.title,p.pub_name as publisher
FROM authors a
join titleauthor ta
join titles t
join publishers p
where ta.au_id = a.au_id and t.title_id = ta.title_id and p.pub_id=t.pub_id
```

# Challenge 2

```python
%%sql
select a.au_id,a.au_lname as last_name,a.au_fname as first_name,t.title,p.pub_name as publisher, count(*) as title_count
FROM authors a
join titleauthor ta
join titles t
join publishers p
where ta.au_id = a.au_id and t.title_id = ta.title_id and p.pub_id=t.pub_id
group by a.au_id,p.pub_name
```

# Challenge 3

```python
%%sql
select a.au_id as AUTHOR_ID,a.au_lname as LAST_NAME,a.au_fname as FIRST_NAME,SUM(s.qty) as TOTAL
FROM authors a
join titleauthor ta
join titles t
join sales s
where ta.au_id = a.au_id and t.title_id = ta.title_id and s.title_id = t.title_id
group by a.au_id
ORDER BY total desc
limit 3
```
