### Challenge 2

Who have published how many at where? Query how many titles each author has published at each publisher. 

· Quién ha publicado cuántos y dónde?

To check if your output is correct, sum up the TITLE COUNT column. The sum number should be the same as the total number of records in Table titleauthor

```python
#Cuántos libros ha escrito cada autor?
```

```python
%%sql
select authors.au_lname as NAME, authors.au_fname as FIRST_NAME, count(*)
from authors join titleauthor on authors.au_id = titleauthor.au_id
            join titles on titleauthor.title_id = titles.title_id
            join publishers on titles.pub_id = publishers.pub_id
group by authors.au_id
```

```python
#Cuántos libros ha escrito cada autor en cada editorial?
```

```python
%%sql
select authors.au_lname as NAME, authors.au_fname as FIRST_NAME, pub_name, count(*)
from authors join titleauthor on authors.au_id = titleauthor.au_id
            join titles on titleauthor.title_id = titles.title_id
            join publishers on titles.pub_id = publishers.pub_id
group by authors.au_id, publishers.pub_name
```

### Challenge 4

Now modify your solution in Challenge 3 so that the output will display all 23 authors instead of the top 3. 

· Modifica tu challenge anterior de manera que el resultado muestre los 23 autores en vez de sólo el top 3

Note that the authors who have sold 0 titles should also appear in your output (ideally display 0 instead of NULL as the TOTAL). Also order your results based on TOTAL from high to low.

```python
%%sql
select authors.au_lname as NAME, authors.au_fname as FIRST_NAME, sum(ytd_sales)
from authors left join titleauthor on authors.au_id = titleauthor.au_id
            left join titles on titleauthor.title_id = titles.title_id
group by authors.au_id
order by sum(ytd_sales) desc
```

```python
%%sql
select authors.au_lname as NAME, authors.au_fname as FIRST_NAME, sum(ifnull(ytd_sales, 0))
from authors left join titleauthor on authors.au_id = titleauthor.au_id
            left join titles on titleauthor.title_id = titles.title_id
group by authors.au_id
order by sum(ytd_sales) desc
```
